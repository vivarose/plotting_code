```python
"""
2019-12-10 Comparing spectrum of fluorinated graphene sample in two different locations.
Both locations are in the region where we applied Ga+ FIB.

@author: Viva R. Horowitz
"""

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime
import seaborn as sns
import numpy as np
import sys
from IPython.display import display

#sns.set(style='ticks')
sns.set_context('poster')

%pylab inline
from matplotlib.mlab import frange
```

```python
datafolder = r"\\cas-fs1.uoregon.edu\Material-Science-Institute\Aleman-Lab\Group\Projects\Fluorinated Graphene\Data\Spectra-OceanOptics"
dimspotfile = "FG_Subt2_16-30-21-897.txt"
brightspotfile= "FG_Subt2_15-48-07-784.txt"

os.chdir(datafolder)


rawdatabright=pd.read_csv(brightspotfile, header=12, sep='\t',
                          index_col=0,
                         names=["Wavelength (nm)","Intensity"])
                            
rawdatadim = pd.read_csv(dimspotfile, header=12, sep='\t',
                         index_col=0,
                        names=["Wavelength (nm)","Intensity"])

display(rawdatabright.tail())
rawdatadim.tail()
```

```python
difference = rawdatabright - rawdatadim
```

```python
difference.plot(ylim=(-400,400),xlim=(500,750),)
```

```python
wind=6
ax = difference.plot(marker='.', markersize=4,linestyle='None',legend=False,color='k',  alpha=.9)
fig = plt.gcf()
plot(pd.DataFrame.rolling(difference, wind, center=True).sum()/wind)
difference.plot(legend=False, linewidth=0.5, alpha=0.2, ax=ax, ylim=(-200,300), xlim=(500,750), color='k')
plt.title("Difference in spectra between bright spot and dim area")
ylabel("Intensity")
```

```python
ax=rawdatabright.plot(marker='.', markersize=4,linestyle='None', legend=False,color='k',  alpha=.9)
fig = plt.gcf()
plot(pd.DataFrame.rolling(rawdatabright, wind, center=True).sum()/wind)
rawdatabright.plot(legend=False, linewidth=0.5, alpha=0.5, ax=ax, ylim=(-120,400), xlim=(500,750),  color='k')
plt.title("Spectrum at bright spot")
ylabel("Intensity")
```

```python
ax = rawdatadim.plot(marker='.', markersize=4,linestyle='None',legend=False,color='k',  alpha=.9)
fig = plt.gcf()
plot(pd.DataFrame.rolling(rawdatadim, wind, center=True).sum()/wind)
rawdatadim.plot(legend=False, linewidth=0.5, alpha=0.5, ax=ax, ylim=(-120,400), xlim=(500,750),  color='k')
plt.title("Spectrum at dim spot")
ylabel("Intensity")
```
